Data Preprocessing

```python
#separating the target and variables
from sklearn.model_selection import train_test_split
```

```python
X=park_data.drop(columns=['name','status'],axis=1)  #for columns axis=1 and row axis=0
```

```python
Y=park_data['status']
```

```python
X
```

```python
Y
```

```python
#splitting the data to train and test the data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
#Split arrays or matrices into random train and test subsets.
#test size=20% so test size is 20% and 80% for training data
#randomstate=2 denotes that it is same as the % used for test size
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler  #to preprocess data and feed to ml
from sklearn import svm  #support vector machine model
from sklearn.metrics import accuracy_score  #evaluate the model
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
print(X_train)
```

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy score of training data:',training_data_accuracy)
```

```python
#accuracy score on testing data
X_test_prediction=model.predict(X_test)
```

```python
testing_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print('Accuracy score of test data:',testing_data_accuracy)
```
